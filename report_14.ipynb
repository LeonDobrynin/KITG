```python
import collections

class Graph:
    def __init__(self, vertices: int) -> None:
        self.V = vertices
        self.graph = []

    def add_edge(self, u: int, v: int, w: int) -> None:
        self.graph.append([u, v, w])

def levit(gr: Graph, start: int = 0):
    m0 = []  
    m1 = collections.deque()  
    m2 = [x for x in range(gr.V)] 

    m1.append(m2.pop(start))  

    dist = [float('inf')] * gr.V  
    dist[start] = 0

    pred = [-1] * gr.V 
    pred[start] = None  

    while m1:
        current = m1.popleft()
        for u, v, w in gr.graph:
            if u == current:
                if v in m2:  
                    dist[v] = dist[u] + w
                    m2.remove(v)
                    m1.append(v)
                    pred[v] = u
                elif v in m1:  
                    if dist[v] >= dist[u] + w:
                        dist[v] = dist[u] + w
                        pred[v] = u
                elif v in m0:  
                    if dist[v] > dist[u] + w:
                        dist[v] = dist[u] + w
                        pred[v] = u
                        m0.remove(v)
                        m1.appendleft(v)
        m0.append(current)

    return dist, pred

# Пример 1: Простой граф
g1 = Graph(5)
g1.add_edge(0, 1, 2)
g1.add_edge(0, 2, 4)
g1.add_edge(1, 2, 1)
g1.add_edge(1, 3, 7)
g1.add_edge(2, 4, 3)

dist1, pred1 = levit(g1, start=0)
print("Расстояния:", dist1)
print("Предшественники:", pred1)

# Пример 2: Граф с отрицательными весами
g2 = Graph(4)
g2.add_edge(0, 1, 1)
g2.add_edge(1, 2, -1)
g2.add_edge(2, 3, 2)
g2.add_edge(3, 0, -2)

dist2, pred2 = levit(g2, start=0)
print("Расстояния:", dist2)
print("Предшественники:", pred2)

# Пример 3: Граф без связей
g3 = Graph(3)
dist3, pred3 = levit(g3, start=0)
print("Расстояния:", dist3)
print("Предшественники:", pred3)
```
